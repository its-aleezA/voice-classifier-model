# BTS Member "Voice" Classifier
This notebook builds a Machine Learning model using Natural Language Processing (NLP) to predict which BTS member a set of lyrics belongs to. We implement Class Balancing (sub-sampling) to ensure the Rap Line doesn't overwhelm the Vocal Line, and N-grams to help the model pick up on short phrases rather than just single words.

```python
import pandas as pd
import numpy as np
import re
import string
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

## 1. Data Loading & Preparation with Sub-Sampling
We will load the individual member CSV files and label them based on the member's name. To prevent the model from just "guessing RM" every time because he has the most lyrics, we will limit the number of songs per member to a maximum of 25. This forces the model to learn the features of the lyrics rather than just the frequency of the labels.

```python
# List of files provided
files = {
    'RM': 'rm.csv',
    'Suga': 'suga.csv',
    'J-Hope': 'jhope.csv',
    'Jimin': 'jimin.csv',
    'V': 'v.csv',
    'Jin': 'jin.csv',
    'Jungkook': 'jungkook.csv'
}

all_data = []

for member, filename in files.items():
    df = pd.read_csv(filename)
    # We only need lyrics
    df = df[['lyrics']].dropna()
    # Assign the member name as the label
    df['member'] = member
    all_data.append(df)

# Combine everything into one big dataframe
df_master = pd.concat(all_data, ignore_index=True)

print(f"Total songs/lyrics snippets: {len(df_master)}")
df_master.head()
```

## 2. Text Preprocessing
Machine Learning models don't understand raw text. We need to clean it by:
1. Converting to lowercase.
2. Removing track annotations (like [Chorus], [Verse 1]).
3. Removing punctuation and special characters.

We'll use a slightly more aggressive cleaning function to remove common "filler" words.

```python
def clean_lyrics(text):
    text = re.sub(r'\[.*?\]', '', text) # Remove [Chorus], etc.
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_master['clean_lyrics'] = df_master['lyrics'].apply(clean_lyrics)
```

## 3. Advanced Vectorization & Splitting
We use TF-IDF (Term Frequency-Inverse Document Frequency). It rewards words that are frequent in one member's lyrics but rare in others (like "hope" for J-Hope or "rockstar" for RM), helping the model differentiate them. Instead of just looking at "love," we now look at "I love," "love you," etc. This captures the "rhythm" of the lyrics better.

```python
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV

# Split 80/20
X_train, X_test, y_train, y_test = train_test_split(
    df_master['clean_lyrics'], 
    df_master['member'], 
    test_size=0.2, 
    random_state=42,
    stratify=df_master['member']
)

# IMPROVEMENT: Use Character N-grams (3 to 6 characters long)
# This helps the model identify "rhythm" and sub-word patterns
tfidf = TfidfVectorizer(
    analyzer='char_wb', 
    ngram_range=(3, 6), 
    max_features=10000
)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print(f"Features created: {X_train_tfidf.shape[1]}")
```

## 4. Model Training & Improved Evaluation
We use the zero_division=0 parameter to handle those warnings you saw earlier.

```python
# LinearSVC doesn't provide probabilities by default, 
# so we wrap it in CalibratedClassifierCV to get the "Confidence Levels"
base_model = LinearSVC(C=1.0, random_state=42, max_iter=2000)
model = CalibratedClassifierCV(base_model) 

model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)

print("--- Improved Classification Report ---")
print(classification_report(y_test, y_pred, zero_division=0))

# Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d',
            xticklabels=model.classes_, yticklabels=model.classes_, cmap='Greens')
plt.title('Updated SVM Model Confusion Matrix')
plt.show()
```

## 5. Live Test Function
Now you can test it with a lyric. Try a lyric from a song like "Winter Bear" or "Daechwita" to see if it works!

```python
def predict_voice_final(text):
    cleaned = clean_lyrics(text)
    vec = tfidf.transform([cleaned])
    pred = model.predict(vec)[0]
    probs = model.predict_proba(vec)[0]
    
    print(f"Lyric snippet: '{text}'")
    print(f"Model Prediction: {pred} ðŸ‘‘")
    print("\n--- Stylistic Confidence ---")
    
    # Sort and print
    results = sorted(zip(model.classes_, probs), key=lambda x: x[1], reverse=True)
    for m, p in results:
        print(f"{m: <10}: {p:.1%}")

# TEST 1: Suga (Daechwita)
# predict_voice_final("I'm a king, I'm a boss")

# TEST 2: Jin (Don't Say You Love Me)
predict_voice_final("Fakin' a smile while we're breakin' apart, I never meant to take it this far")
```
